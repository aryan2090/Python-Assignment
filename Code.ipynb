```python
#3
import csv
rows=[]
with open("per_capita_income.csv", 'r') as csvfile:
    csvreader=csv.reader(csvfile)
    for row in csvreader: 
        rows.append(row)
    t=1
    sum=0
    count=0
    for row in rows:
        for col in row:
            try:
                if t>3 and t%2==0:
                    sum+=int(col)
                    t+=1
                    count+=1
                else:
                    t+=1
            except:
                sum+=int(float(col))
                t+=1
                count+=1
    print("Count is:",count)#Just to know count
    print("The mean is:",sum/count)
```

```python
#4
import csv
rows=[]
with open("per_capita_income.csv", 'r') as csvfile:
    csvreader=csv.reader(csvfile)
    for row in csvreader: 
        rows.append(row)
    n=184
    count=0
    mid1=0
    mid2=0
    for row in rows[1:]:
        count+=1
        if count==n/2:
            mid1=int(float(row[1]))
        elif count==((n/2)+1):
            mid2=int(float(row[1]))
    print(count)
    print("The median is:",(mid1+mid2)/2)

```

```python
#5
import csv
import statistics
rows=[]
l=[]
with open("per_capita_income.csv", 'r') as csvfile:
    csvreader=csv.reader(csvfile)
    for row in csvreader: 
        rows.append(row)
    rows[0].append("Income per capita//1000")
    for row in rows[1:]:
        try:
            row.append(int(row[1])//1000)
        except:
            row.append(int(float(row[1]))//1000)
for row in rows:
    print(row)
    l.append(row[2])
print("The mode of the new column is:",statistics.mode(l))
```
